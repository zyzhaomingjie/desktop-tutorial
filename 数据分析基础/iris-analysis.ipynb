```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris_type = iris['Name'].unique()
colors = ['#c72e29','#098154','#fb832d']
for i in range(len(iris_type)):
    plt.scatter(iris.loc[iris['Name'] == iris_type[i], 'SepalLength'],
                iris.loc[iris['Name'] == iris_type[i], 'SepalWidth'],
                s = 50,
                c = colors[i],
                label = iris_type[i])
    plt.ylabel('SepalWidth')
    plt.xlabel('SepalLength')
plt.legend(loc = 'upper right')
```

```python
for i in range(len(iris_type)):
    plt.scatter(iris.loc[iris['Name'] == iris_type[i], 'PetalLength'],
                iris.loc[iris['Name'] == iris_type[i], 'PetalWidth'],
                s = 50,
                c = colors[i],
                label = iris_type[i])
    plt.ylabel('PetalWidth')
    plt.xlabel('PetalLength')
plt.legend(loc = 'upper left')
```
